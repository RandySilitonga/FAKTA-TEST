```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Load dataset
data_path = Path("data/training/combined.csv")
if data_path.exists():
    df = pd.read_csv(data_path)
    print(f"Dataset loaded: {len(df)} samples")
else:
    print("Dataset not found. Run: python src/data/collect.py")
```

```python
# Dataset overview
print(df.info())
print("\nLabel distribution:")
print(df['label'].value_counts())
```

```python
# Visualize label distribution
plt.figure(figsize=(8, 5))
df['label'].value_counts().plot(kind='bar', color=['#dc3545', '#28a745', '#ffc107'])
plt.title('Label Distribution')
plt.ylabel('Count')
plt.show()
```

```python
# Text length analysis
df['text_length'] = df['text'].astype(str).str.len()
df['word_count'] = df['text'].astype(str).str.split().str.len()

print(f"Text length stats:")
print(df['text_length'].describe())
print(f"\nWord count stats:")
print(df['word_count'].describe())
```

```python
# Text length by label
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='label', y='text_length')
plt.title('Text Length by Label')
plt.show()
```

```python
# Source distribution
plt.figure(figsize=(10, 5))
df['source'].value_counts().plot(kind='barh')
plt.title('Data Sources')
plt.xlabel('Count')
plt.show()
```

```python
# Sample texts per label
for label in df['label'].unique():
    print(f"\n{'='*50}")
    print(f"Label: {label}")
    print(f"{'='*50}")
    sample = df[df['label'] == label]['text'].iloc[0]
    print(sample[:300] + "..." if len(sample) > 300 else sample)
```
